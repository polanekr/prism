```python
import plotly.graph_objects as go
import numpy as np

def plot_3d_interactive(volume, z_coords, y_coords, x_coords, level=50, opacity=0.3):
    """
    Interaktív 3D izodózis felület (Isosurface) Plotly-val.
    
    Paraméterek:
    - level: Izodózis szint %-ban (pl. 50 -> az 50%-os dózis felületét rajzolja).
    - opacity: Átlátszóság (0-1).
    """
    
    max_d = np.max(volume)
    isovalue = max_d * (level / 100.0)
    
    # Meshgrid a Plotly-nak (figyelj az indexelésre!)
    # A Plotly (x, y, z) koordinátákat vár flattened formában
    Z, Y, X = np.meshgrid(z_coords, y_coords, x_coords, indexing='ij')
    
    fig = go.Figure(data=go.Isosurface(
        x=X.flatten(),
        y=Y.flatten(),
        z=Z.flatten(),
        value=volume.flatten(),
        isomin=isovalue,
        isomax=max_d,
        surface_count=5, # Hány réteget rajzoljon a level és a max között
        opacity=opacity,
        caps=dict(x_show=False, y_show=False),
        colorscale='Jet',
        colorbar=dict(title='Dose [Gy]')
    ))
    
    fig.update_layout(
        title=f"3D Interactive Dose Volume (>{level}%)",
        scene=dict(
            xaxis_title='X [mm]',
            yaxis_title='Y [mm]',
            zaxis_title='Depth Z [mm]',
            zaxis=dict(autorange='reversed') # Z tengely fordítása
        )
    )
    
    fig.show(renderer='browser')

# Használat:
# plot_3d_interactive(vol, z, y, x, level=30)
```

```python
data = np.load('/home/polanekr/Kutatás/PRISM-Framework/examples/results/films/PDD1.npz') # Vagy a te fájlneved
vol = data['volume']
        
# Koordináták kezelése (ha nincsenek elmentve, generáljuk)
if 'z_coords' in data:
    z = data['z_coords']
    y = data['y_coords']
    x = data['x_coords']
```

```python
plot_3d_interactive(vol, z, y, x, level=30)
```
